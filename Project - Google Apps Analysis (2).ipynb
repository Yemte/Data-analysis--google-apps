```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apps = pd.read_csv('googleplaystore.csv')
```

```python
apps
```

Group the data by category and explore the average rating

```python
gb = apps.groupby('Category')

cat_mean = gb.mean()

cat_mean
```

```python
rating = cat_mean.plot(kind='bar', y='Rating', figsize=(20,6))

#rating.set_ylim(3, 4.5)

rating.axhline(y=4, color='r')
```

We can see that Events apps are usually rated very high in the market. Other highly rated categories are Art, Books and Education.

All categories except Dating apps have an average rating above 4. It could either be that people are often frustrated with dating apps, or that these apps tend to have poor design and functionality.

```python
#Dtata cleaning

def replaceSize(x):
    if x.endswith('M'):
        return float(x[:-1])
    elif x.endswith('k'):
        return float(x[:-1])/1000
    else:
        return np.nan
 
apps.Size = apps.Size.apply(replaceSize)
apps
```

```python
apps.Installs = apps.Installs.str.replace('+', '')
apps.Installs = apps.Installs.str.replace(',', '')
apps.Installs = apps.Installs.astype(float)
apps
```

```python
apps.Price = apps.Price.str.replace('$','0')
apps.Price = apps.Price.astype(float)
```

```python
# apps[apps.App.duplicated(keep=False)]
apps.drop_duplicates(subset='App', keep='last', inplace=True)
apps
```

```python
# - Visualize the distribution of price for paid apps only
paid_app =apps[apps['Price']>0]
k=paid_app.groupby('Category')
fig, ax = plt.subplots(figsize=(12,6))
price_dis = k.Category.value_counts().plot(kind='bar', rot=90, color='g',ax=ax)

price_dis.set_title('number of paid apps in each category')
```

```python
# Plot the Content Rating column
colors = plt.cm.gist_rainbow(apps.Rating.value_counts().values)
r=apps.Rating.value_counts().plot(kind='bar', rot=90, color=colors,figsize=(20,10))
r.set_title('Number of apps by rating',fontsize =16,color = "red")
```

```python
# - Do free apps have more installs than paid apps on average?
gb_price = apps.groupby(apps.Price > 0).mean()

pa = gb_price.plot(kind='bar', y='Installs', figsize=(10,10), rot=0,color='orange')

pa.set_title('Paid vs Free apps comparison', fontsize=15)
pa.set_xlabel('')
pa.set_xticklabels(['Free apps', 'Paid apps'], {'fontsize':12})

```
